```python
import pandas as pd
```

```python
ventas = pd.read_csv('ventas_50.csv')
ventas
```

**Calcular total del producto: Cantidad x Precio**

```python
ventas['total'] = ventas.Cantidad * ventas.Precio
ventas.head()
```

**Filtrar productos con total > 5**

```python
mayor_a_5 = ventas[ventas.total > 5]
mayor_a_5.head()
```

**Calcular el total general de ventas**

```python
total_general = ventas.total.sum()
```

```python
'Total General de Ventas = {:.2f}'.format(total_general)
```
